```python
import mlflow
import pandas as pd
import matplotlib.pyplot as plt

import os
#os.environ["PATH"] = "/home/users/dmoreno2016/.TinyTeX/bin/x86_64-linux:" + os.environ["PATH"]

#plt.rcParams.update({
#    "text.usetex": True,
#    "pgf.texsystem": "pdflatex",
#    "font.family": "serif",
#    "font.serif": ["Palatino"],
#})


import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import classification_report
```

```python
mlflow.set_tracking_uri("file:./ATAT/results/ml-runs")
```

```python
data_name =  'ztf_ff'
experiment_name = f"classification/{data_name}/hp_tuning"

experiment = mlflow.get_experiment_by_name(experiment_name)
if experiment:
    experiment_id = experiment.experiment_id
    print(f"Experiment ID for '{experiment_name}': {experiment_id}")
else:
    print(f"Experiment '{experiment_name}' not found.")

runs = mlflow.search_runs(experiment_ids=[experiment_id])
runs = runs[
    (runs.status == 'FINISHED') & 
    (runs['metrics.mix/f1s_test/None_days'].notna())
    ]

display(runs.head())
print(f'{runs.shape}\n')

for col in runs.columns:
    print(col)
```

```python
columns_hp = [
    'params.lc/num_encoders', 'params.lc/embedding_size', 'params.lc/embedding_size_sub',
    'params.ft/num_encoders', 'params.ft/embedding_size', 'params.ft/embedding_size_sub',
    ]
```

```python
fold_run = {run_id: (fold, exp_name) for run_id, fold, exp_name in zip(
    runs['run_id'], runs['params.general/fold'], runs['params.general/experiment_name']
    )}
fold_run
```

```python
import numpy as np

# Diccionario para almacenar las métricas por exp_name
metrics = {}

for run_id, (fold, exp_name) in fold_run.items():
    # Cargar predicciones
    path_clf_report = f'./ATAT/results/ml-runs/{experiment_id}/{run_id}/artifacts'
    atat_df_predictions = pd.read_parquet(f'{path_clf_report}/predictions_test.parquet').set_index('id')
    y_test = atat_df_predictions['y_true'].values
    atat_y_pred = atat_df_predictions['y_pred'].values
    
    # Calcular el reporte de clasificación
    report_atat = classification_report(y_test, atat_y_pred, output_dict=True)
    macro_precision = report_atat['macro avg']['precision']
    macro_recall = report_atat['macro avg']['recall']
    macro_f1 = report_atat['macro avg']['f1-score']

    hp_values = [runs[runs.run_id == run_id][col].values[0] for col in columns_hp]
    
    # Guardar los valores en la lista correspondiente a exp_name
    if exp_name not in metrics:
        metrics[exp_name] = {'precision': [], 'recall': [], 'f1-score': [], 'hp_values': []}
    metrics[exp_name]['precision'].append(macro_precision*100)
    metrics[exp_name]['recall'].append(macro_recall*100)
    metrics[exp_name]['f1-score'].append(macro_f1*100)
    metrics[exp_name]['hp_values'].append(hp_values)

# Crear DataFrame con los valores medios y desviaciones estándar
rows = []
for exp_name, values in metrics.items():
    row_macro = [exp_name]
    row_macro.extend([
        f"{np.mean(values['precision']):.2f} ± {np.std(values['precision']):.2f}",
        f"{np.mean(values['recall']):.2f} ± {np.std(values['recall']):.2f}",
        f"{np.mean(values['f1-score']):.2f} ± {np.std(values['f1-score']):.2f}"
    ])
    row_macro.extend(values['hp_values'][0])
    rows.append(row_macro)

# Crear DataFrame
columns = ['exp_name', 'macro_precision', 'macro_recall', 'macro_f1'] + columns_hp
report_macro_df = pd.DataFrame(rows, columns=columns).set_index('exp_name')
report_macro_df = report_macro_df.sort_values(by='macro_f1', ascending=False)
report_macro_df
```

## HBRF Hyperparameter Tuning

```python
import pandas as pd

objects_partitions = pd.read_parquet('../../data_acquisition/ztf_forced_photometry/preprocessed/partitions/241209_ndetge8/partitions.parquet')
partitions_test = objects_partitions[objects_partitions['partition'] == 'test'].set_index('oid')
y_test = partitions_test['class_name'].values
y_test
```

```python
import os
import numpy as np
import pandas as pd
import yaml
from sklearn.metrics import f1_score, classification_report

# Ruta base donde están los trials
parent_dir = "classifiers/results/241209_ndetge8/HBRF_20250304-225053"

# Listar los trials
trials = sorted([d for d in os.listdir(parent_dir) if d.startswith("trial_")])

# Diccionario para almacenar las métricas por trial
metrics = {}

for trial in trials:
    trial_path = os.path.join(parent_dir, trial)
    
    # Cargar hiperparámetros del trial
    hparams_file = os.path.join(trial_path, "hparams.yaml")
    with open(hparams_file, "r") as file:
        hparams = yaml.safe_load(file)

    # Lista para almacenar los F1-scores de los folds de este trial
    precision_scores = []
    recall_scores = []
    f1_scores = []

    # Iterar sobre los folds dentro del trial
    folds = sorted([d for d in os.listdir(trial_path) if d.startswith("fold_")])
    for fold in folds:
        fold_path = os.path.join(trial_path, fold)
        pred_file = os.path.join(fold_path, "predictions_test.parquet")

        if os.path.exists(pred_file):
            # Cargar predicciones
            hbrf_df_predictions = pd.read_parquet(pred_file)
            hbrf_df_predictions = hbrf_df_predictions[hbrf_df_predictions['shorten'] == 'None']
            hbrf_df_predictions.index = hbrf_df_predictions.index.astype(str).str.split('_').str[1]
            hbrf_df_predictions = hbrf_df_predictions.reindex(partitions_test.index) 
            hbrf_y_pred = hbrf_df_predictions['y_pred'].values

            # Calcular reporte de clasificación
            report = classification_report(y_test, hbrf_y_pred, output_dict=True)
            macro_precision = report['macro avg']['precision']
            macro_recall = report['macro avg']['recall']
            macro_f1 = report['macro avg']['f1-score']

            precision_scores.append(macro_precision * 100)
            recall_scores.append(macro_recall * 100)
            f1_scores.append(macro_f1 * 100)

    # Guardar los resultados en el diccionario
    if trial not in metrics:
        metrics[trial] = {'precision': [], 'recall': [], 'f1-score': [], 'hp_values': []}
    
    metrics[trial]['precision'].extend(precision_scores)
    metrics[trial]['recall'].extend(recall_scores)
    metrics[trial]['f1-score'].extend(f1_scores)
    metrics[trial]['hp_values'].append([hparams[key] for key in hparams if key not in ["mean_f1", "std_f1", "trial"]])

# Crear DataFrame con los valores medios y desviaciones estándar
rows = []
for trial, values in metrics.items():
    row = [trial]
    row.extend([
        f"{np.mean(values['precision']):.2f} ± {np.std(values['precision']):.2f}",
        f"{np.mean(values['recall']):.2f} ± {np.std(values['recall']):.2f}",
        f"{np.mean(values['f1-score']):.2f} ± {np.std(values['f1-score']):.2f}"
    ])
    row.extend(values['hp_values'][0])  # Asumiendo que los hiperparámetros son los mismos para todos los folds
    rows.append(row)

# Crear DataFrame final
columns_hp = list(hparams.keys())
columns = ['trial', 'macro_precision', 'macro_recall', 'macro_f1'] + columns_hp
report_macro_df = pd.DataFrame(rows, columns=columns).set_index('trial')
report_macro_df = report_macro_df.sort_values(by='macro_f1', ascending=False)
columns_to_remove = ['n_jobs', 'verbose']
report_macro_df = report_macro_df.drop(columns=[col for col in columns_to_remove if col in report_macro_df.columns], errors='ignore')
report_macro_df
```

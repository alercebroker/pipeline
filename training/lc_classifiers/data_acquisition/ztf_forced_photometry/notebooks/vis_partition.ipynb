```python
import pandas as pd
```

```python
partitions_new_tax = pd.read_parquet('preprocessed/partitions/241209_ndetge8_20folds/partitions.parquet')
partitions_new_tax
```

```python
# Filtrar los datos de test y convertirlos en un diccionario
ids_test_new_tax = set(partitions_new_tax[partitions_new_tax["partition"] == "test"].oid.values)
print(f'Test {len(ids_test_new_tax)}')

# Calcular las proporciones de train y val en cada fold
num_folds = 20 
for fold in range(num_folds):
    num_ids_train = partitions_new_tax[partitions_new_tax.partition == f'training_{fold}'].shape[0]
    num_ids_val = partitions_new_tax[partitions_new_tax.partition == f'validation_{fold}'].shape[0]
    total = num_ids_train + num_ids_val

    print(f'Fold {fold}: train {num_ids_train} ({(num_ids_train/total) * 100 :.1f}%) - val {num_ids_val} ({(num_ids_val/total) * 100:.1f} %)')
```

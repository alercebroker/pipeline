```python
import numpy as np
import matplotlib.pyplot as plt


# --- PASO 2: CONFIGURACIN DEL GRFICO ---

# Define los nombres para cada canal. Puedes personalizarlos!
# Basado en tu cdigo anterior y el orden comn, podra ser algo as:
channel_names = [
    "Canal 0: Science",
    "Canal 1: Difference",
    "Canal 2: Template/Reference",
    "Canal 3: Mscara Science", # Suposicin, ya que es el 4to canal que agregaste
    "Canal 4: Mscara Difference",
    "Canal 5: Mscara Template/Reference"
]

# Obtenemos el nmero de canales directamente del shape del array
num_channels = stamps_rubin.shape[3]

# Creamos una figura con subplots. Una cuadrcula de 2x3 es ideal para 6 canales.
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
# Aplanamos el array de 'axes' para iterar fcilmente con un solo ndice (de 0 a 5)
axes = axes.flatten()


# --- PASO 3: ITERAR Y GRAFICAR CADA CANAL ---

print("\nGenerando grficos para cada canal...")
for i in range(num_channels):
    ax = axes[i]
    
    print(f"  Procesando {channel_names[i]}...")

    # Extraer todos los pxeles para el canal 'i' de cada dataset
    # La sintaxis [:, :, :, i] selecciona TODOS los elementos de las primeras 3 dimensiones,
    # pero solo el ndice 'i' de la cuarta dimensin (canales).
    # .flatten() convierte el resultado en un array 1D para el histograma.
    rubin_pixels = stamps_rubin[:, :, :, i].flatten()
    dp1_pixels = stamps_dp1[:, :, :, i].flatten()
    ztf_pixels = stamps_ztf[:, :, :, i].flatten()
    
    # --- Calcular el rango comn para el eje X (idntico a tu lgica anterior) ---
    
    # 1. Combinar datos para calcular un rango representativo de todos los datasets
    combined_data = np.concatenate((rubin_pixels, dp1_pixels, ztf_pixels))
    
    # 2. Eliminar valores infinitos o NaN para que los clculos no fallen
    combined_data = combined_data[np.isfinite(combined_data)]
    
    # Si no quedan datos vlidos para este canal, saltamos al siguiente
    if combined_data.size == 0:
        ax.set_title(f"{channel_names[i]}\n(Sin datos vlidos)", fontsize=14)
        continue

    # 3. Calcular percentiles para definir lmites y evitar outliers extremos
    q_low = np.quantile(combined_data, 0.01)
    q_high = np.quantile(combined_data, 0.99)
    
    # Asegurar que el rango no sea cero si los datos son muy homogneos
    if q_low >= q_high:
        q_low -= 1e-6 # Pequeo ajuste para evitar errores
        q_high += 1e-6

    # --- Graficar los histogramas para el canal actual ---
    ax.hist(rubin_pixels, bins=100, density=True, alpha=0.7, label='Rubin (v0.0.1)')#, range=(q_low, q_high))
    ax.hist(dp1_pixels, bins=100, density=True, alpha=0.7, label='DP1 (v0.0.4)')#, range=(q_low, q_high))
    ax.hist(ztf_pixels, bins=100, density=True, alpha=0.7, label='ZTF')#, range=(q_low, q_high))
    
    # --- Configuracin esttica del subplot ---
    ax.set_title(f"Distribucin de Pxeles - {channel_names[i]}", fontsize=14)
    ax.set_xlabel("Valor del Pxel")
    ax.set_ylabel("Densidad")
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)
    ax.set_yscale('log') # Usar escala logartmica es muy til para ver detalles en las colas

# Si hay ms subplots que canales (p. ej. 8 subplots para 7 canales), ocultamos los vacos.
for j in range(num_channels, len(axes)):
    axes[j].set_visible(False)
    
plt.tight_layout() # Ajusta los subplots para que no se superpongan
plt.show()

print("\nGrficos generados exitosamente!")
```

```python
# Nombres de los canales de mscara y sus ndices
mask_channels = {
    "Canal 3: Mscara Science": 3,
    "Canal 4: Mscara Difference": 4,
    "Canal 5: Mscara Template/Reference": 5
}

datasets = {
    "Rubin": stamps_rubin,
    "DP1": stamps_dp1,
    "ZTF": stamps_ztf
}

# Diccionario para almacenar los resultados
mask_stats = {}

print("\n--- Estadsticas de Pxeles Enmascarados (valor=1) ---")

for name, data in datasets.items():
    mask_stats[name] = {}
    print(f"\nProcesando Dataset: {name}")
    
    for mask_name, channel_idx in mask_channels.items():
        # Extraer el canal de la mscara
        mask_channel = data[:, :, :, channel_idx]
        
        # Calcular estadsticas. La media de un array de 0s y 1s es la proporcin de 1s.
        # np.mean() es mucho ms eficiente que np.sum() / array.size
        proportion_masked = np.mean(mask_channel)
        percentage_masked = proportion_masked * 100
        
        mask_stats[name][mask_name] = percentage_masked
        
        print(f"  - {mask_name}: {percentage_masked:.2f}% de los pxeles estn enmascarados.")

# Convertir a un DataFrame de pandas para una visualizacin ms bonita (opcional pero recomendado)
stats_df = pd.DataFrame(mask_stats)
print("\n--- Tabla Resumen de Porcentajes (%) ---")
print(stats_df)


# --- PASO 3: VISUALIZACIN CON GRFICO DE BARRAS AGRUPADO ---

# Usaremos los datos del DataFrame para graficar
labels = stats_df.index
rubin_values = stats_df['Rubin']
dp1_values = stats_df['DP1']
ztf_values = stats_df['ZTF']

x = np.arange(len(labels))  # las localizaciones de las etiquetas
width = 0.25  # el ancho de las barras

fig, ax = plt.subplots(figsize=(9, 5))

# Creacin de las barras para cada dataset
rects1 = ax.bar(x - width, rubin_values, width, label='Rubin (v0.0.1)')
rects2 = ax.bar(x, dp1_values, width, label='DP1 (v0.0.4)')
rects3 = ax.bar(x + width, ztf_values, width, label='ZTF')

# Aadir texto, ttulos y etiquetas
ax.set_ylabel('Porcentaje de Pxeles Enmascarados (%)', fontsize=12)
ax.set_title('Proporcin de Pxeles Enmascarados por Tipo de Imagen y Telescopio', fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Funcin para aadir etiquetas de valor encima de cada barra
def autolabel(rects):
    """Aade una etiqueta con el valor encima de cada barra en rects."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 puntos de offset vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast 


# --- PASO 2: FUNCIN SIMPLIFICADA PARA EXTRAER Y APLANAR LOS PXELES ---
def extract_pixel_values(df, column_name):
    """
    Toma un dataframe y el nombre de una columna que contiene arrays de NumPy.
    Devuelve un nico array de numpy con todos los valores de los pxeles.
    """
    # Filtra filas donde la columna podra tener valores nulos
    valid_stamps = df[column_name].dropna()
    
    # Si no hay datos, devuelve un array vaco
    if valid_stamps.empty:
        return np.array([])
        
    # Usa np.concatenate para unir de forma eficiente todos los arrays de imgenes
    # aplanados en uno solo.
    all_pixels = np.concatenate([img.flatten() for img in valid_stamps])
    
    return all_pixels

# --- PASO 3: EXTRACCIN DE TODOS LOS DATOS ---
print("\nExtrayendo valores de pxeles...")

# Datos de Rubin
rubin_science_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Science_data')
rubin_diff_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Difference_data')
rubin_template_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Template_data')


# Datos de dp1
dp1_science_pixels = extract_pixel_values(stamps_dp1_df, 'visit_image')
dp1_diff_pixels = extract_pixel_values(stamps_dp1_df, 'difference_image')
dp1_template_pixels = extract_pixel_values(stamps_dp1_df, 'reference_image')

# Datos de ZTF
ztf_science_pixels = extract_pixel_values(stamps_ztf_df, 'science')
ztf_diff_pixels = extract_pixel_values(stamps_ztf_df, 'diff')
ztf_template_pixels = extract_pixel_values(stamps_ztf_df, 'reference')

print("Extraccin completada.")
```

```python

# --- PASO 4: CREACIN DE GRFICOS CON LMITES AUTOMTICOS ---

plot_data = [
    {
        "title": "Distribucin de Pxeles - Science",
        "rubin_data": rubin_science_pixels,
        "dp1_data": dp1_science_pixels,
        "ztf_data": ztf_science_pixels
    },
    {
        "title": "Distribucin de Pxeles - Difference",
        "rubin_data": rubin_diff_pixels,
        "dp1_data": dp1_diff_pixels,
        "ztf_data": ztf_diff_pixels
    },
    {
        "title": "Distribucin de Pxeles - Template/Reference",
        "rubin_data": rubin_template_pixels,
        "dp1_data": dp1_template_pixels,
        "ztf_data": ztf_template_pixels
    }
]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, data in enumerate(plot_data):
    ax = axes[i]
    
    # --- INICIO DE LA MODIFICACIN ---
    
    # 1. Combinar datos de Rubin y ZTF para calcular un rango comn
    combined_data = np.concatenate((data["rubin_data"], data["ztf_data"]))
    
    # 2. Eliminar valores infinitos o NaN, que destrozan los clculos
    combined_data = combined_data[np.isfinite(combined_data)]
    
    # 3. Calcular el percentil 1 y 99 para definir los lmites
    # Puedes ajustar estos valores (ej. 0.5 y 99.5) si todava hay muchos outliers
    q_low = np.quantile(combined_data, 0.01)  # Lmite inferior
    q_high = np.quantile(combined_data, 0.99) # Lmite superior

    # Asegurarse de que el rango no sea un solo punto si los datos son muy homogneos
    if q_low == q_high:
        q_low -= 1
        q_high += 1

    print(f"Grfico '{data['title']}': Rango ajustado a [{q_low:.2f}, {q_high:.2f}]")

    # 4. Usar el parmetro 'range' en la funcin hist
    ax.hist(data["rubin_data"], bins=100, density=True, alpha=0.7, label='Rubin (v0.0.1)', range=(q_low, q_high))
    ax.hist(data["dp1_data"], bins=100, density=True, alpha=0.7, label='DP1 (v0.0.4)', range=(q_low, q_high))
    ax.hist(data["ztf_data"], bins=100, density=True, alpha=0.7, label='ZTF', range=(q_low, q_high))

    # --- FIN DE LA MODIFICACIN ---
    
    ax.set_title(data["title"], fontsize=14)
    ax.set_xlabel("Valor del Pxel")
    ax.set_ylabel("Densidad")
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

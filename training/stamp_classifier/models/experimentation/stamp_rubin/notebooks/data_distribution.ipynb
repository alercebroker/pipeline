```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast 


# --- PASO 2: FUNCIÃ“N SIMPLIFICADA PARA EXTRAER Y APLANAR LOS PÃXELES ---
def extract_pixel_values(df, column_name):
    """
    Toma un dataframe y el nombre de una columna que contiene arrays de NumPy.
    Devuelve un Ãºnico array de numpy con todos los valores de los pÃ­xeles.
    """
    # Filtra filas donde la columna podrÃ­a tener valores nulos
    valid_stamps = df[column_name].dropna()
    
    # Si no hay datos, devuelve un array vacÃ­o
    if valid_stamps.empty:
        return np.array([])
        
    # Usa np.concatenate para unir de forma eficiente todos los arrays de imÃ¡genes
    # aplanados en uno solo.
    all_pixels = np.concatenate([img.flatten() for img in valid_stamps])
    
    return all_pixels

# --- PASO 3: EXTRACCIÃ“N DE TODOS LOS DATOS ---
print("\nExtrayendo valores de pÃ­xeles...")

# Datos de Rubin
rubin_science_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Science_data')
rubin_diff_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Difference_data')
rubin_template_pixels = extract_pixel_values(stamps_rubin_df, 'flux_Template_data')


# Datos de dp1
dp1_science_pixels = extract_pixel_values(stamps_dp1_df, 'visit_image')
dp1_diff_pixels = extract_pixel_values(stamps_dp1_df, 'difference_image')
dp1_template_pixels = extract_pixel_values(stamps_dp1_df, 'reference_image')

# Datos de ZTF
ztf_science_pixels = extract_pixel_values(stamps_ztf_df, 'science')
ztf_diff_pixels = extract_pixel_values(stamps_ztf_df, 'diff')
ztf_template_pixels = extract_pixel_values(stamps_ztf_df, 'reference')

print("ExtracciÃ³n completada.")
```

```python

# --- PASO 4: CREACIÃ“N DE GRÃFICOS CON LÃMITES AUTOMÃTICOS ---

plot_data = [
    {
        "title": "DistribuciÃ³n de PÃ­xeles - Science",
        "rubin_data": rubin_science_pixels,
        "dp1_data": dp1_science_pixels,
        "ztf_data": ztf_science_pixels
    },
    {
        "title": "DistribuciÃ³n de PÃ­xeles - Difference",
        "rubin_data": rubin_diff_pixels,
        "dp1_data": dp1_diff_pixels,
        "ztf_data": ztf_diff_pixels
    },
    {
        "title": "DistribuciÃ³n de PÃ­xeles - Template/Reference",
        "rubin_data": rubin_template_pixels,
        "dp1_data": dp1_template_pixels,
        "ztf_data": ztf_template_pixels
    }
]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, data in enumerate(plot_data):
    ax = axes[i]
    
    # --- INICIO DE LA MODIFICACIÃ“N ---
    
    # 1. Combinar datos de Rubin y ZTF para calcular un rango comÃºn
    combined_data = np.concatenate((data["rubin_data"], data["ztf_data"]))
    
    # 2. Eliminar valores infinitos o NaN, que destrozan los cÃ¡lculos
    combined_data = combined_data[np.isfinite(combined_data)]
    
    # 3. Calcular el percentil 1 y 99 para definir los lÃ­mites
    # Puedes ajustar estos valores (ej. 0.5 y 99.5) si todavÃ­a hay muchos outliers
    q_low = np.quantile(combined_data, 0.01)  # LÃ­mite inferior
    q_high = np.quantile(combined_data, 0.99) # LÃ­mite superior

    # Asegurarse de que el rango no sea un solo punto si los datos son muy homogÃ©neos
    if q_low == q_high:
        q_low -= 1
        q_high += 1

    print(f"GrÃ¡fico '{data['title']}': Rango ajustado a [{q_low:.2f}, {q_high:.2f}]")

    # 4. Usar el parÃ¡metro 'range' en la funciÃ³n hist
    ax.hist(data["rubin_data"], bins=100, density=True, alpha=0.7, label='Rubin (v0.0.1)', range=(q_low, q_high))
    ax.hist(data["dp1_data"], bins=100, density=True, alpha=0.7, label='DP1 (v0.0.4)', range=(q_low, q_high))
    ax.hist(data["ztf_data"], bins=100, density=True, alpha=0.7, label='ZTF', range=(q_low, q_high))

    # --- FIN DE LA MODIFICACIÃ“N ---
    
    ax.set_title(data["title"], fontsize=14)
    ax.set_xlabel("Valor del PÃ­xel")
    ax.set_ylabel("Densidad")
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
